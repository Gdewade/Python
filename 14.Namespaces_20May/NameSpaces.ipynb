# NameSpaces

### 1.Local namespace : variable/identifire define inside the function and access only inside the function

```python
def greet():
    msg="Hello" # this is local namspace
    print(msg)

greet()
```

### 2. Global namspace : variable/identifire define outside the function and access everywhere in program

```python
lang='Python'
version =3.11 # global namspace

print(lang)

def show():
    print(version)
show()

```

## Shadowing : means overlap local and global var

```python
count=100

def show_count():
    count=999
    print("Inside:",count)
show_count()
print("Outside",count)
```

### Global keyword : it dosent create new var it just replace outside global var value with local var value

```python
count=100

def update_count():
    global count
    count=999
    print("Inside:",count)
update_count()
print("Outside",count)
```

### globals() and locals() : used to return dictionaries of namespaces

```python
city='paris'
country='france'
population=2_200_000

def city_info():

    district='marais'
    global population
    population += 50_000

    print("Inside locals() : ",locals())
    print("Outside globals() : ",globals())

print("Module locals() :",locals())
print("Module globals :",globals())

city_info()
```

```python
# used of globals() and locals()

jan_salary=1500
feb_salary=2000
march_salary=4000

months=['jan_salary','feb_salary','march_salary']

for i in months:

    print(i,'=',globals()[i]) # i is key and globals() gives us corresponding values

#globals()

```

```python
def add(x,y): return x+y
def multiply(x,y): return x*y
def subtract(x,y): return x-y

operations=['add','multiply','subtract']

for op in operations:

    result=globals()[op](10,2)
    print(f'{op}(10,2)={result}')

print(globals())
```

## Nested function / Inner function : the function define in another function

```python
def order_summary():
    
    item='coffee'          
    price=4.50
    print('----order summary-----')

    def show_receipt():

        tax= price * 0.10
        total= price+tax
        print(f'Item: {item}')
        print(f'Price: ${price:.2f}')
        print(f'Tax : ${tax:.2f}')
        print(f'Total : ${total:.2f}')

    show_receipt()

order_summary()
                        
```

### 3. Enclosive scope : variable declare in outer function and accessing in only outer and inner function

```python
def order_summary():
    
    item='coffee'          # Enclosive scope
    price=4.50             # Enclosive scope
    print('----order summary-----')

    def show_receipt():

        tax= price * 0.10
        total= price+tax
        print(f'Item: {item}')
        print(f'Price: ${price:.2f}')
        print(f'Tax : ${tax:.2f}')
        print(f'Total : ${total:.2f}')

    show_receipt()

order_summary()
```

#### Flow of namspace in python

```python
country='japan' # Global

def outer_fn():
    city='Tokyo' # Enclosing

    def inner_fn():
        district = 'shibuva' # Local
        print("Flow:",district,city,country)
        print(len(district)) # Built-in

    inner_fn()

outer_fn()
```

```python
# same name of var but scopes are different 

def outer():
    score = 45           # outer's score
    print('outer score:', score)
    print('outer id:   ', id(score))

    def inner():
        score = 90       # inner's OWN score — different!
        print('inner score:', score)
        print('inner id:   ', id(score))

    inner()

outer()
```

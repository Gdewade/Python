```python
# Q1.  Write a program that demonstrates all four LEGB levels
#      in one nested structure:

x='global'

def outer():
    x='enclosing'
    def inner():
        x='local'
        print(x)    # this is local because variable name of enclose and local is same but local overlaps on enclose here
    inner()
outer()             # this is global
print(x)
```

```python
# Q2.  Repeat Q1 but remove x = 'local' from inner().
#      Now which x does inner() print? Explain the LEGB
#      lookup path Python follows.

# it will print enclosing x because enlcosing namspace is access if local is not there 

x='global'
def outer():
    x='enclosing'
    def inner():
        print(x)
    inner()
outer()
print(x)

# python used LEGB path for maintain the sequence of scopes
# L(local) : it define inside inner function and accessable only in inner function
# E(enclosing) : it define inside outer function and accessable till outer function call
# G(global) : it define top level of code and it accessable everywere in the program
# B(built-in) : if scopes are not found then python checks its own built-in or system names
```

```python
# Q11. Write the full payroll program from 10.py:

base_wage = 15.0    # Global 

def payroll(hr):
    overtime_rate=1.5   # Enclosing

    def compute_pay():

        regular=min(hr,40)*base_wage                # Locals
        overtime=max(hr-40,0)
        ot_pay=overtime*base_wage*overtime_rate
        total= regular+ ot_pay

        print(f'Hours:    {hr}')
        print(f'Regular:  ${regular:.2f}')
        print(f'OT Pay:   ${ot_pay:.2f}')
        print(f'Total:    ${total:.2f}')

    compute_pay()

payroll(50)
```

```python
# Q28. Write the dynamic function call program from 7.py:

def add(x,y):return x+y
def mul(x,y):return x*y
def sub(x,y):return x-y

operations=['add','mul','sub']
for op in operations:
    result=globals()[op](10,3)
    print(f'{op}(10,3)={result}')

# globals()[op](10,3) = it takses operations from list using op variable and performing on values 10,3
```

# Set Properties

### 1. Mutability = after creating set u can add,remove or change elements 

### 2. Conversion = convert from any data type to set

```python
#string to set

string1="Hello"
set1=set(string1)
print(set1)
```

```python
#list to set

list1=[1,2,3,4]
set1=set(list1)
print(set1)
```

```python
#tuple to set

tupl11=(1,2,3,4,5)
set1=set(tupl11)
print(set1)
```

```python
#Dictonary to set

dict1={"Name:":"Gayatri","Surname:":"Dewade"}
set1=set(dict1)
print(set1)
```

```python
#range to set

set1=set(range(1,11))
print(set1)
```

### 3 .Aliasing = multiple variables pointing to the same mem location

```python
set1={1,2,3}
print(id(set1))
set2=set1
print(set2)
print(id(set2))
```

### 4. Cloning = multiple variables pointing to the different mem location using copy()

```python
set1={1,2,3}
print(id(set1))
set2=set1.copy()
print(set2)
print(id(set2))
```

### 5. Searching/ Membership operator = in

```python
set1={1,2,3}
print(2 in set1)
```

### 6. Identity = is , its check that 2 var are from same mem location

```python
set1={1,2,3}
print(id(set1))

set2=set1

print(set2)
print(id(set2))

set3={30,40,50}

print(set1 is set2)
print(set3 is set1)
```

### 7 .Comparision

# Built in funtions in set

### 1. len() -  returns count of items in a set

```python
set1={10,20,30}
print(len(set1))
```

### 2 . max()

```python
set1={10,20,30}
print(max(set1))
```

### 3. min()

```python
set1={10,20,30,False}
print(min(set1))
```

### 4. sum()

```python
set1={10,20,30,True}
print(sum(set1))
```

### 5. any()

```python
set1={0,0,0,True}
print(any(set1))
```

### 6 . all()

```python
set1={1,2,3}
print(all(set1))
```

### 7. sorted() - returns new sorted ascending list

```python
set1={10,4,2,88,100}
print(sorted(set1))
```

```python
set1={10,4,2,88,100}
print(reversed(set1))
```

# Methods in set

### 1. add()

```python
set1={1,2,3}
set1.add(4)
print(set1)
```

### 2. update() - add elements of another set in existing set

```python
set1={1,2,3}
set2={4,5,6}
set3={"Gayatri",8.1,9+3j,True}

set1.update(set2,set3)
print(set1)
```

### 3 . copy() - create an independent copy of the set

```python
set1={10,20,30,40,50,60,70}
set2=set1.copy()

print(set1)
print(id(set1))

print(set2)
print(id(set2))

```

### 4. remove() - remove the specific element

```python
set1={20,30,40,50}
set1.remove(50)
print(set1)
```

```python
set1.remove(60)
```

### 5. discard()

```python
set1={20,30,40,50}
set1.discard(50)
print(set1)
```

```python
set1.discard(60)
```

### 6 . clear()

```python
set1={20,30,40,50}
set1.clear()
print(set1)
```

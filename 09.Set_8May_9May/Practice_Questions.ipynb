```python
# Q10. Create set1 = {1, 2, 3} and set2 = set1.
#      i)  Print id(set1) and id(set2). Are they the same?
#      ii) Now create set3 = set1.copy()
#          Print id(set1) and id(set3). Are they the same? Why not?

set1={1,2,3}
print(id(set1))

set2=set1 # here we used aliasing and in aliasing memory address of both variables are same
print(id(set2))

set3=set1.copy() # here we used cloning and in cloning memory address of both variables are different
print(id(set3))

```

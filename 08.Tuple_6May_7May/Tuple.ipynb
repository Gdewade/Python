# Tuple ()

## Properties Of Tuple

### 1. Homogeneous tuple = store elements of same data type

```python
tuple1=(10,30,40,50)
print(tuple1)
print(type(tuple1))
```

### 2. Heterogeneous tuple = store elements of different data type

```python
tuple2=("Gayatri",1,0.1,3+2j,True)
print(tuple2)
```

### 3. Ordered = elements keeps their insertion order,access calues by index value

```python
# maintain insertion order 

tuple3=(1,2,3,4)
print(tuple3)
```

### 4. Duplication are allow

```python
tuple3=(1,2,3,4,3)
print(tuple3)
```

### 5. Immutable in nature

```python
tuple4=(1,2,3,4)
print(id(tuple4))

tuple4[0]=0 # comes error bcoz we cant change,add,remove elements
```

```python
tuple4.append(1)
```

### 6. Empty tuple = we can create empty tuple

```python
tuple3=()
print(tuple3)
```

### 7. Single item tuple = if a tuple has only one item a trailing comma is required inside the ()

```python
tuple4=(1)
print(tuple4)
print(type(tuple4)) # they doesnt show () in output

tuple4=(1,)
print(tuple4)
print(type(tuple4))
```

### Accessing elements

```python
tup1=(10,20,30,40,50)
print(tup1)

#+ve indexing
print(tup1[0])

#-ve indexing
print(tup1[-5])

#slicing
print(tup1[0:3])

print(tup1[0],tup1[2],tup1[4])
print(tup1[0:5:2])

print(tup1[0:])
print(tup1[:5])
print(tup1[:])
print(tup1[::-1]) # for reverse tuple
print(tup1[::-2])




```

# Loops = iterating a tuple
### 1. while loop

```python
my_tuple=(10,20,30)
index=0
while (index<len(my_tuple)):
    print(my_tuple[index])
    index=index+1
```

### 2. for loop

```python
my_tuple=(10,20,30)

for i in my_tuple:
    print(i)
```

```python
my_tuple=(10,20,30)

for i,j in enumerate(my_tuple):
    print(i,j)
```

### 2. Concatenation / Merging  = Joining tuple

```python
tup=(1,2,3)
tup1=(4,5,6)
tup3=tup+tup1
print("Concat of tup",tup3)
```

```python
tup=(1,2,3)
tup2=(4,5,6)
tup3=(7,8,9)
tup4=tup+tup2+tup3
print(tup4)
print(type(tup4))
```

### 3 . Converting into tuple
##### string,list,set,range

```python
string1="Gayatri"
print(tuple(string1))
```

```python
# converting mutable to immutable means list to tuple
list=[1,2,3,4,5]
print(tuple(list))
```

```python
set={0,1,2,3,4}
print(tuple(set))
```

```python
print(tuple(range(5)))
```

### 5. Aliasing : multiple variables pointing to the same tuple in mem

```python
tup1=(10,20,30)
#tup2=(40,50,60)
tup2=tup1
print(tup1)
print(id(tup1))
print(tup2)
print(id(tup2))
```

### 6. Cloning : multiple variavbles pointing to the diff tuple in mem

```python
tup1=(10,20,30)
print(id(tup1))

tup2=tuple(tup1)
print(tup2)
print(id(tup2))
```

### 7. Searching - in operator / membership operator = used to check whether particular element is present in a tuple or not

```python
tuple1=("Gayatri",31,3+2j)
print(3+2j in tuple1)
```

### 8. Identity - 'is' to check if two var refer to the same mem location

```python
tup1=(10,20,30)
tup2=tup1

print(tup1 is tup2)
```

```python
tup3=(10,20,30)
print(tup2 is tup3)
```

### 9. Emptiness - 'not' check if tuple is empty

```python
tup1=()
if not tup1:
    print("Tuple is empty")
else:
    print("Tuple is not empty")
```

```python
print(not tup1)
```

```python
tup1=(1,)
if not tup1:
    print("Tuple is empty")
else:
    print("Tuple is not empty")
```

```python
print(not tup1)
```

### 10. Comparision = comparing tuple contents element by element 

```python
a=(1,2,3,4)
b=(1,2,5)

print(a<b)
print(a>b)
print(a==b)
print(a>=b)
print(a<=b)
```

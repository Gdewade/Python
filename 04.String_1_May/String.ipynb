# String Slicing

### It extract substring or part of the string


#### Syntax :
#### string_name( start_index:end_index:step)
#### step = used to skip character from String 

```python
print(str1[0:7]) #by default step = 1

print(str1[:7])  #by default start = 0

print(str1[0:])  #by default end = length of the string or end_index+1

print(str[:])

print(str[1:5])

print(str[0:1])
```

```python
print(str1[0:7:1])
print(str1[0:7:2])
print(str1[0:7:4])
```

### 2 . Concatenation +

```python
#to join 2 strings 
greet="Good"
greet2="Morning"
msg=greet+" "+greet2
print(msg)
```

### 3 . Duplication *

```python
#to repeat same string multiple times
msg=greet+" "+greet2+" "+" "
print(msg * 3)

#we can't used sep here bcoz here we used single string we cant separate single string for sep we need multiple or 2 string
```

### 4. Membership Testing
###    check if substring is exist in that string or not (using 'in' or 'not in')

```python
text="i am from pune"

print("pune" in text)
print("mumbai" in text)

print("am" not in text)
print("we" not in text)
```

### 2 . String Search method 
### find() returns lowest index of substring or if not found they show -1

```python
text = "Hello, welcome to python programming !"

index=text.find("python")
print(index)

index=text.find("Java")
print(index)
```

### replace() rplace old variable with new

```python
text = "Hello, welcome to python programming !"

new_text=text.replace("python","Java") 
print(new_text)
```

### lstrip() = remove left whitespace, rstrip() = remoce right whitespaces, strip() = remove both whitespaces

```python
text=" Hello World "

result=text.lstrip()
print(result)

print(type(result))
```

```python
text=" Hello World "

result=text.rstrip()
print(result)
```

```python
text=" Hello World "

result=text.strip()
print(result)
```

### 3. Spliting Methods

```python
text= "A,B,C-D-E"

result=text.split("-")
print(result)
```

### partition() = divides string in 3 parts

```python
text= "Hello, welcome to python programming !"
result=text.partition("to")
print(result)
```

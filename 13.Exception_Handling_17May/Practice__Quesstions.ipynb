```python
# Q20. Write a safe_double(value) function that:
#      - Uses isinstance() to check if value is int or float
#      - If yes: returns value * 2
#      - If no: raises TypeError with message
#        "Expected int or float, got <type>"
#      Test with 5, 3.14, and '5'.

def safe_double(value):
    if isinstance(value,(int,float)):
        return value * 2
    else:
        raise TypeError (f"Expected int or float, got {type(value)}.__name__")
    
try:
    print(safe_double(5))
    print(safe_double(3.14))
    print(safe_double('5'))

except TypeError as err:
    print("Caught an error : ",err)
```

```python
# Q30. Write a parse_age(text) function that:
#      - Converts text to int (handles ValueError)
#      - Validates that age is between 0 and 120
#      - Returns the age if valid, else returns None with a message
#      Test with '25', 'abc', and '200'.

def parse_age(text):
    try:
        age = int(text)
        if 0 <= age <= 120:
            return age
        else:
            print(f"Age {age} is out of valid range (0-120).")
            return None
    except ValueError:
        print(f"'{text}' is not a valid number.")
        return None

print(f"Result: {parse_age('25')}\n")
print(f"Result: {parse_age('abc')}\n")
print(f"Result: {parse_age('200')}\n")
```

```python
# Q41. Write a program that handles ZeroDivisionError,
#      TypeError, and ValueError in the same try block
#      using separate except clauses.
#      Test all three cases.


def test(a,b):
    try:

        # cause ValueError
        num1=int(a)
        num2=int(b)


        # cause TypeError
        if num1==999:
            result=num1+"string"
        
        # cause ZeroDivisionError
        result = num1/num2
        return result
    
    except ValueError as err:
        print("ValueError : ",err)

    except TypeError as err:
        print("TypeError : ",err)

    except ZeroDivisionError as err:
        print("ZeroDivisionerr : ",err)

# Test Case 1: Testing ValueError
print("Test 1 ('abc', 5):", test("abc", 5))

# Test Case 2: Testing ZeroDivisionError
print("Test 2 (10, 0):", test(10, 0))

# Test Case 3: Testing TypeError
print("Test 3 (999, 5):", test(999, 5))

# Test Case 4: Testing Success Case
print("Success")
print("Test 4 (10, 2):", test(10, 2))
    
```

```python
# Q42. Write a compute_log(x) function that:
#      - Handles x = 0 (math.log(0) → ValueError)
#      - Handles x < 0 (negative → ValueError)
#      - Returns math.log(x) for valid positive x
#      Test with 100, 0, and -5.

import math

def compute_log(x):
    try:
        return math.log(x)
    except ValueError as err :
        print("ValueError : ",err)
    
print("x = 100\n",compute_log(100))
print("x = 0",compute_log(0))
print("x = -5",compute_log(-5))
```

```python
# Q43. Write a complete program that:
#      i)   Takes a string from a variable (simulate user input)
#      ii)  Converts it to int (handles ValueError)
#      iii) Divides 100 by that number (handles ZeroDivisionError)
#      iv)  Prints the result if successful
#      Test with '5', '0', and 'abc'.

test=['5','0','abc']

for simulated_input in test:
    print(f"\n Testing with input : '{simulated_input}'")

    try:
        user_input=simulated_input
        number=int(user_input)
        result=100/number

    except ValueError as err:
        print("ValueError : ",err)

    except ZeroDivisionError as err:
        print("ZeroDivisionError : ",err)

    else:
        print("Result is successful")

```

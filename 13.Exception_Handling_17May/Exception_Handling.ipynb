### 2. OverflowError :- Floating pt arithmetic result is to large to be represented in Python

```python
import math

math.exp(1000)
```

```python
math.exp(10)
```

```python
import math
import sys

# Example 1: Float overflow — e^1000 is beyond float's max range
print("=== Example 1: math.exp() Overflow ===")
try:
    huge = math.exp(1000)    # e^1000 ≈ 5×10^434 — too large for float!
    print(huge)
except OverflowError as err:
    print('OverflowError:', err) 
```

```python
# Example 2: Python integers NEVER overflow — they just grow!
print("\n=== Example 2: Python int — No Overflow! ===")
big_int = 10 ** 500             # 10 to the power 500 — no error!
print('Big int digits:', len(str(big_int))) 
```

```python
# Example 3: Float multiplication overflow
print("\n=== Example 3: Float Multiplication Overflow ===")
try:
    x = 1.8e308 * 10          # 1.8×10^308 × 10 = 1.8×10^309 — beyond float max
    print(x)
except OverflowError as err:
    print('Float overflow:', err)

    print("====== check ======")
if math.isinf(x):           # here we raised exception for inf which is in x variable
    print("Float overflow detected:", x)
else:
    print(x)

# Show float's actual maximum
print("\n=== Float Info ===")
print(f"Max float value: {sys.float_info.max:.2e}")
```

```python
# ── OverflowError — Prevention with decimal module ────────────────────────────
# The 'decimal' module provides arbitrary precision — no overflow for large numbers
from decimal import Decimal

# Using Decimal instead of float for precision-critical calculations
print("=== Using decimal module for large numbers ===")
a = Decimal('1.8') * Decimal('10') ** 308    # No overflow!
print(f"Result computed safely using Decimal (first 30 digits): {str(a)[:30]}...")
```

```python
# Safe exponential with guard
def safe_exp(x, max_val=700):
    """Safe wrapper for math.exp — prevents OverflowError."""
    if x > max_val:
        return float('inf')    # Return infinity as a safe fallback
    return math.exp(x)

print("\n=== Safe Exponential ===")
print(f"safe_exp(10)   = {safe_exp(10):.4f}")
print(f"safe_exp(1000) = {safe_exp(1000)}")   # Returns inf instead of crashing
```

### 4. ValueError : A functiom receives an argument of the correct type but the value is invalid for the operation

```python
import math
```

```python
# Example 1 : Convert non-numeric string into int

# Problem 

try:
    score=int('eighty')
except ValueError as err:
    print('ValueError: ',err) # base 10 means by default int have base 10
```

```python
# Solution 1 : 

def to_int_1(s):
    try:
        return int(s)
    except ValueError:
        print("Invalid Input type")
        return None
     
print(to_int_1('eighty'))
print(to_int_1('42'))
```

```python
# Solution 2

def to_int_2(s1):
    if not s1.isdigit():
        print(f"'{s1}' is not a valid age")
        return None
    return int(s1)

print(to_int_2('eighty'))
print(to_int_2('42'))
```

```python
# Example 2 : Square root of -ve numbers

try:
    print(math.sqrt(-4))
except ValueError as err:
    print("ValueError: ",err)
```

```python
# Solution 

def safe_sqrt(x):
    if x < 0:
        print(f"Cant compute square root of -ve number : {x}")
        return None
    return math.sqrt(x)

print(safe_sqrt(25))
print(safe_sqrt(-25))
```

```python
# Q11. Given the dictionary:
#      dict1 = {"name": "Akshay", "roll_no": 100,
#               "weight": 65.6, "placement": True}
#      Write three separate for loops to:
#      i)  Print all keys
#      ii) Print all values
#      iii) Print both keys and values together

dict1={"name":"Akshay","roll_no":100,"weight":65.6,"placement":True}

print("Print all keys","\n")
for i in dict1.keys():
    print(i)

print("\n","Print all values","\n")
for j in dict1.values():
    print(j)

print("\n","Print both","\n")
for a,b in dict1.items():
    print(a,"=",b)
```

```python
# Q18. Create dict1 = {"Name": "Akshay", "roll_no": 101}
#      and assign dict2 = dict1.
#      i)  Print id(dict1) and id(dict2). Are they the same?
#      ii) Change dict1["Name"] to "Rahul".
#          Print dict2. Has it changed? Why?

# i)
dict1 = {"Name": "Akshay", "roll_no": 101}
dict2=dict1
print(id(dict1))
print(id(dict2))

# ii)
dict1["Name"]="Rahul"
print(dict2)

#Yes, its change because here we use aliasing , it means we refer two variables to the same memory location
# and any modification done can affect on new one also


```

```python
# Q20. What is the difference between aliasing and cloning?
#      Write code to show both and use id() to prove they differ.

#aliasing , it means we refer two variables to the same memory location
# and any modification done can affect on new one also
#coloning its also like aliasing but here modifications are not affect on eachother

print("Aliasing")
dict1 = {"Name": "Akshay", "roll_no": 101}
dict2=dict1
print(id(dict1))
print(id(dict2))

print("\n")

print("Cloning")
dict1 = {"Name": "Akshay", "roll_no": 101}
dict2=dict1.copy()
print(id(dict1))
print(id(dict2))

```

```python
# Q38. What is the difference between pop(), popitem(), and clear()?
#      Write one example for each.

# i) pop() =  used to remove particular key with value
data = {40: 'a', 30: 'b', 60: 'c'}
data.pop(30)
print(data)

# ii) popitem() = used to remove last key-pair from dictonary
data = {40: 'a', 30: 'b', 60: 'c'}
data.popitem()
print(data)

# iii) clear() = used to remove all elements from the dictonary
data = {40: 'a', 30: 'b', 60: 'c'}
data.clear()
print(data)
```

# Looping in Dictonaries
### 1. Iterate over key-value pairs Syntax: for key,value in dict_name.items():

```python
dict1={"Name":"Gayatri","name":"Vedant","Age":21,"Roll_no":21,"City":"Pune","Placement":True}
for i,j in dict1.items():
    print("Key:",i,"         Value :",j)
```

### 2. Iterate over key only Syntax : for key/i/j in dict_name.keys():

```python
dict3={"Name":"Gayatri"}
for i in dict3.keys():
    print(i)
```

### 3. Iterate over value only syntax :  for value/i/j in dict_name.values():

```python
dict3={"Name":"Gayatri"}
for i in dict3.values():
    print(i)
```

# enumerate()

```python
dict3={"Name":"Gayatri","Age":21}

for i,(k,v) in enumerate(dict3.items()): # difference is var_name.items() 
    print(i,k,v)
```

 # Operations in Dict

### 1 . Mutable - we can change,remove,replace key value pairs

```python
dict1={"Name":"Gayatri","Age":21,"Roll_no":21,"City":"Pune","Placement":True}
dict1["Name"]= "Avishkar"
print(dict1)
```

### 2. Concatenation - used '|' operator

```python
dict_a={"Name":"Gayatri"}
dict_b={"Surname":"Dewade"}

dict_c=dict_a | dict_b

print(dict_c)
```

```python
dict_a={"Name":"Gayatri"}
dict_b={"Surname":"Dewade"}

dict_c=dict_a | dict_b | {"Age":21}

print(dict_c)
```

### 3. Convert into dict - dict()

```python
#Tuple

tuple1=(('a',1),('b',2),('c',3))
print(dict(tuple1))
```

```python
print(tuple1[0])
```

```python
#list

list1=[['a',1],['b',2],['c',3]]
print(dict(list1))
```

```python
#set

set1={{'a',1},{'b',2},{'c',3}}
print(dict(set1))
```

```python
# to assign same value for all keys
dict1=dict()
set2={'a','b','c'}
for key in set2:
    print(key)
    dict1[key]=2

print(dict1)
```

```python
# to assign different value for all keys
dict1=dict()
count=0
set2={'a','b','c'}
for key in set2:
    print(key)
    dict1[key]=count
    count=count+1

print(dict1)
```

```python
#string

string1="Gayatri"
dict1={}

for i in string1:
    dict1[i]=10

print(dict1)
```

```python
#range

my_range=range(5)
dict1={} # empty dict

for i in my_range:
    dict1[i]=10

print(dict1)
```

### 4. Aliasing

```python
dict1={"Name":"Gayatri","Surname":"Dewade"}
dict2=dict1
print(id(dict1))
print(id(dict2))
```

```python
dict1["Name"]="Vedant"

print(dict2)
```

### 5. Cloning

```python
dict1={"Name":"Gayatri","Surname":"Dewade"}
dict2=dict1.copy()
print(id(dict1))
print(id(dict2))
```

```python
dict1["Name"]="Vedant"

print(dict2)
```

### 6. 'in' searching 

### to check keys

```python
dict1={"Name":"Gayatri","Age":21,"Roll_no":21,"City":"Pune","Placement":True}
print("Name" in dict1)
print("Surname" in dict1)
```

### to check values

```python
dict1 = {"name": "Akshay", "roll_no": 100}
print("Akshay" in dict1.values())
```

### 7. 'is' identity operator

```python
dict1={"Name":"Gayatri","Surname":"Dewade"}
dict2=dict1
print(id(dict1))
print(id(dict2))

print(dict1 is dict2)
```

```python
dict1={"Name":"Gayatri","Surname":"Dewade"}
dict2=dict1.copy()
print(id(dict1))
print(id(dict2))

print(dict1 is dict2)
```

# Methods in Dictonary

### 1. update() - used to add or update multiple key-value pair

```python
dict1={'a':2}
dict1.update({'b':3})
print(dict1)
```

```python
dict1={'a':2}
dict2={'b':3}
dict1.update(dict2)

print(dict1)
```

```python
dict1["b"]=4

print(dict1)
```

### 2. pop(key) - to delete pair

```python
dict3={'a': 2, 'b': 3}
dict3.pop("a")

print(dict3)
```

### 3. popitem() - to delete last pair of dict

```python
dict3={'a': 2, 'b': 3}
dict3.popitem()

print(dict3)
```

### 4. clear()

```python
dict3={'a': 2, 'b': 3}
dict3.clear()

print(dict3)
```

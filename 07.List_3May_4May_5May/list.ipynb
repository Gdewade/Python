# Looping through list
#### to process each element of the list individually
##### 1. while loop

```python
print("Program starts")

list4 = [10, 20, 30, 40, 50]

index = 0
while (index < len(list4)):
    print(list4[index])   
    index = index + 1

print("Program ends")
```

```python
print("Program starts")

list4 = [10, 20, 30, 40, 50]

index = -1
while (index > -6):
    print(list4[index])   
    index = index - 1

print("Program ends")
```

##### 2 . for loop

```python
list4 = [10, 20, 30, 40, 50]

for i in list4:
    print(i)
```

```python
for i in [10, 20, 30, 40, 50]:
    print(i)
```

### 4. Aliasing / shallow copy 
#### Aliasing occurs when multiple variables refer to the same list object in memory
#### ( means multiple variable store same memory location )
#### when one var assign to another variable they store in same mem location

```python
list10=[1,2,3]
print("Before Aliasing list10 :",list10)
print(id(list10))
print("\n")

list9=[4,5,6]
print("Before Aliasing list9 :",list9)
print(id(list9))
print("\n")

list10=list9 # Aliasing
print("After Aliasing list10=list9 :",list10)
print(id(list10))
print(id(list9))
```

#### Disadvantage : Changes to one list it will affect on other list also

```python
list9[0]=9
print(list9)
```

```python
print(list10)
print(id(list10))
```

### 5. cloning / Deep copy
#### It creates separate copy of a list on diff mem location

#### Advantage : Change to one list doesnt affect to another

```python
list11=[10,20,30]
print(id(list11))
print(list11)
print("\n")

list12=[40,50,60]
print(id(list12))
print(list12)
print("\n")

list12=list11.copy()
print(id(list12))
print(list12)

```

```python
list11[0]=90
print(list11)

print(list12)

# here we see Change to one list doesnt affect to another
```

# Methods of list
#### Syntax : list_name.method_name()

### 1 . Addinig methods
#### -append(value) = adds a single item in to the end of the list, its modify original list
#### -insert(index,value) = insert item on a particular index
#### -extend(list_name) = its add all elements from another list to the end of the current list



```python
#append()

list1=[10,20,30]
print(id(list1))

list1.append(40)
print(list1)

print(id(list1))
```

```python
#insert()

list2=[10,20,30,40,50]
print(list2)

list2.insert(4,45)
print(list2)

list2.insert(0,5)
print(list2)

list2.insert(0,"Numbers from 10 to 50:")
print(list2)

# list2.insert(10,60) # when we gives index out of bound it will send that value on just next position of ending element
# print(list2)
# print(list2[10])
```

```python
#extend()

list3=[10,20,30]
print(list3)

list3.extend(list4)
print(list3)

list3.append(list4) # it creates nested list
print(list3)
```

### 2 . Removing methods

#### -pop() = it removes last item from list and when we provide index number it will remove the item from that index

```python
#remove

list5=[2,4,5,6,8,10,2]
list5.remove(5)
print(list5)
```

```python
#pop()

list5=[2,4,5,6,8,10,2]

list5.pop(2)
print(list5)
```

```python
# Q12. What will be the output? How does while loop work with a list using index?

# index is used to check condition in while loop

colors = ["Red", "Green", "Blue", "Yellow"]
idx = 0
while idx < len(colors):
    print(colors[idx])
    idx = idx + 1
print("Done")
```

```python
# Q13. What will be the output? What does enumerate() return?

#enumerate() returns index with value

languages = ["Python", "Java", "C++", "JavaScript"]
for index, value in enumerate(languages):
    print(index, value)
```

```python
# Q14. What will be the output? What is the difference between for loop and enumerate()?

#for loop is a repetation controll loop in python used when we know the value of iteration for traversing in loop
#enumerate() is a built in function in python used in list to print index with values

departments = ["HR", "IT", "Finance", "Marketing"]
print("-- Using for loop --") # it gives values only
for dept in departments:
    print(dept)

print("-- Using enumerate --") #it gives values with index
for idx, dept in enumerate(departments):
    print(idx, dept)
```

```python
# Q19. What will be the output?

grades = [55, 72, 48, 90, 65, 38, 80]
for idx, val in enumerate(grades):
    if val >= 50:
        print(f"Index {idx}: {val} -- Pass")
    else:
        print(f"Index {idx}: {val} -- Fail")
```

```python
# Q20. What will be the output?

inventory = ["Pen", "Pencil", "Eraser", "Sharpener", "Notebook"]
print("First item:", inventory[0])
print("Last item:", inventory[-1])
print("Middle items:", inventory[1:4])
print("Every alternate:", inventory[::2])

for pos, prod in enumerate(inventory):
    print(pos, "-->", prod)
```

```python
# Q19. Write a class Rectangle with:
#      - __init__ setting: length, width
#      - Method area(self) that returns length * width
#      - Method perimeter(self) that returns 2*(length+width)
#      Create two Rectangle objects with different sizes.
#      Print area and perimeter for both.


class Rectangle:

    def __init__(self,length,width):

        self.length=length
        self.width=width
    
    def area(self):
        return (self.length * self.width)
    
    def parameter(self):
        return 2*(self.length + self.width)

val1=Rectangle(3,2)
print("Area(3,2):",val1.area())
print("Parameter(3,2):",val1.parameter())
print('\n')
val2=Rectangle(6,8)
print("Area(6,8):",val2.area())
print("Parameter(6,8):",val2.parameter())

    
```

```python
# Q26. Write the complete BankATM class from class.py.
#      Identify in each method:
#      i)   Where self appears in the method definition
#      ii)  Where self is used to access attributes
#      iii) Where self is used to call another method
#      Label all three places with comments.

print("start program")
class BankATM:

    # ── Constructor ──────────────────────────────────────
    def __init__(self):           # self in method definition
        self.secret_code = ''     # Attribute 1: PIN
        self.account_bal  = 0     # Attribute 2: Balance
        self.show_menu()          # Auto-show menu! , self used to call another method

    # ── Method 1: Menu & Routing ─────────────────────────
    def show_menu(self):          # self in method definition
        choice = input('''
                  Welcome! Choose an option:
                  1. Create PIN
                  2. Change PIN
                  3. Check Balance
                  4. Withdraw Cash
                  5. Exit
                  Your choice: ''')
        if   choice == '1':
            self.set_pin()      # self used to call another method
        elif choice == '2':
            self.update_pin()   # self used to call another method
        elif choice == '3':
            self.check_balance()    # self used to call another method
        elif choice == '4':
            self.do_withdraw()  # self used to call another method
        elif choice == '5':
            exit()

    # ── Method 2: Create PIN ─────────────────────────────
    def set_pin(self):             # self in method definition
        new_code          = input('Enter new PIN: ')
        self.secret_code  = new_code        # self is used to access attributes
        opening           = input('Opening balance: ')
        self.account_bal  = int(opening)    # self is used to access attributes
        print('PIN created!')
        self.show_menu()    # self used to call another method

    # ── Method 3: Change PIN ─────────────────────────────
    def update_pin(self):           # self in method definition
        current = input('Current PIN: ')
        if current == self.secret_code:     # self is used to access attributes
            fresh            = input('New PIN: ')
            self.secret_code = fresh        # self is used to access attributes
            print('PIN updated!')
        else:
            print('Wrong PIN!')
        self.show_menu()        # self used to call another method

    # ── Method 4: Check Balance ──────────────────────────
    def check_balance(self):         # self in method definition
        entered = input('PIN: ')
        if entered == self.secret_code:     # self is used to access attributes
            print('Balance:', self.account_bal) # self is used to access attributes
        else:
            print('Wrong PIN!')
        self.show_menu()        # self used to call another method

    # ── Method 5: Withdraw ───────────────────────────────
    def do_withdraw(self):          # self in method definition
        entered = input('PIN: ')
        if entered == self.secret_code:     # self is used to access attributes
            amt = int(input('Amount: '))
            if amt <= self.account_bal:  # Balance Guard!       # self is used to access attributes
                self.account_bal -= amt     # self is used to access attributes
                print('Done! New bal:', self.account_bal)
            else:
                print('Insufficient!')
        else:
            print('Wrong PIN!')
        self.show_menu()        # self used to call another method

    # ── Launch Application ───────────────────────────────────
obj1 = BankATM()  # One line starts EVERYTHING!
print("end program")
```

```python
# Q3. BASIC - Static Method
#    Create a class MathHelper with:
#    - @staticmethod square(n)   → returns n * n
#    - @staticmethod cube(n)     → returns n * n * n
#    - @staticmethod is_even(n)  → True if n is even

#    Call all 3 methods using class name (no object).

class MathHelper:

    @staticmethod
    def square(n):
        return n*n
    
    @staticmethod
    def cube(n):
        return n*n*n
    
    @staticmethod
    def is_even(n):
        if n % 2 == 0:
            return True
        else:
            return False

print(MathHelper.square(2))
print(MathHelper.cube(2))
print(MathHelper.is_even(2))
```

```python
# Q6. MEDIUM - Static Variable + Static Method
#    Create a BankMachine class with:
#    - static variable cust_counter = 1
#    - __init__ assigns cust_id from counter
#      and increments counter
#    - instance method display() → prints cust_id
#    - @staticmethod get_total_customers() →
#      returns cust_counter - 1

#    Create 3 objects, call display() on each,
#    print total customers using class name.

class BankMachine:

    cust_counter = 1

    def __init__(self):
        self.cust_id = BankMachine.cust_counter
        BankMachine.cust_counter +=1

    def display(self):
        print("Customer id :",self.cust_id)

    @staticmethod
    def get_total_customers():
        return BankMachine.cust_counter - 1

b1=BankMachine()
b1.display()

b2=BankMachine()
b2.display()

b3=BankMachine()
b3.display()

print(BankMachine.get_total_customers())

    
```

```python
# Q7. MEDIUM - Pin Validator (Static Method)
#    Create a BankMachine class with:
#    - @staticmethod validate_pin_format(pin_str):
#      True if all digits AND length is exactly 4

#    Test with:
#    "4523" → True
#    "ab12" → False
#    "12"   → False
#    "12345"→ False
#    "0000" → True

class BankMachine:

    @staticmethod
    def validate_pin_format(pin_str):
        if pin_str.isdigit() and len(pin_str) == 4:
            return True
        else:
            return False
        
print(BankMachine.validate_pin_format("4523"))
print(BankMachine.validate_pin_format("ab12"))
print(BankMachine.validate_pin_format("12"))
print(BankMachine.validate_pin_format("12345"))
print(BankMachine.validate_pin_format("0000"))
```

```python
# Q8. MEDIUM - Instance vs Static Variable Behaviour
#    Create a class Employee with:
#    - static variable company_name = "TechCorp"
#    - instance variables: emp_name, salary
#    - method show() → prints company_name,
#      emp_name, salary

#    Create 2 Employee objects and call show().
#    Now change Employee.company_name to "NewCorp".
#    Call show() again on both objects.
#    Observe: company_name changes for ALL objects.
#    Write what you observe and why.


class Employee:

    company_name="TechCorp"

    def __init__(self,emp_name,salary):
        self.emp_name=emp_name
        self.salary=salary

    def show(self):
        print("Employee Name :",self.emp_name)
        print("Salary :",self.salary)
        print(Employee.company_name)
        print("-----------------------------")

e1=Employee("Gayatri",50000)
e1.show()

e2=Employee("Vedant",60000)
e2.show()

Employee.company_name="NewCorp"

e1=Employee("Gayatri",50000)
e1.show()

e2=Employee("Vedant",60000)
e2.show()

# yes company name is changes for all objects
# Because its a static variable and its same for all object after changing its value also
```

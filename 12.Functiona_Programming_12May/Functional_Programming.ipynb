# 1. Lambda Function

### Syntax :- lambda arguments : expression

```python
# Regular Function

def sqr(n):
    return n**2

print(sqr(2))
#or
result=sqr(3)
print(result)
```

```python
# Lambda Function

sqr=lambda n : n**2

print(sqr(2))
#or
result=sqr(3)
print(result)
```

```python
check= lambda n : n>0
print("Its Positive :",check(2))
print("Its Positive :",check(-2))
```

```python
compare= lambda a,b : a>b
print(compare(5,6))
```

```python
concat=lambda a,b : a+" "+b
print(concat("Gayatri","Dewade"))
```

# 2.  Map Function

### Syntax :- var_name = map(function_name,iterable)

```python
# get cubes of values

values=[1,2,3,4,5,6] # ----> Iterable

def cube(n):         # ----> Function
    return n**3 

result=map(cube,values)
print(list(result))  # -----> Object convert into iterable(list,tuple)
```

#### Using map + lambda 

```python
values=[1,2,3,4,5,6]
double = map(lambda n : n **3 , values) 
print(list(double))
```

```python
# get string length

fruits=["Apple","Banana","Mango"]
lengths=map(lambda f : len(f) , fruits)
print(list(lengths))
```

```python
# add 18 % gst on cost

costs=[100,200,670,800]
tax=map(lambda i : i*1.18 ,costs)
print(list(tax))
```

# 3 . Filter Function 

### Syntax :- var_name = filter(function,iterable) but it returns True and False

```python
# only returns marks > 50

marks=[42,48,99,75,33,80]
result=filter(lambda n: n >=50,marks)
print(list(result)) # filter the input and give output according to condition
```

```python
marks=[42,48,99,75,33,80]
result=map(lambda n: n >=50,marks)
print(list(result)) # same no. of input and output 
```

```python
# filter for length > 5

fruits=["Apple","Banana","Mango"]
result=filter(lambda n : len(n)>5,fruits)
print(list(result))
```

```python
# adults only age > =18

ages=[12,25,17,30,15,22]
adults=filter(lambda n : n>=18,ages)
print(list(adults))
```

# 4 . Reduce Function
### here we used 2 variable for looping = i,j

```python
from functools import reduce

# addition of numbers

num=[1,2,3,4,5]
addition=reduce(lambda i,j : i+j ,num)
print(addition)
```

```python
from functools import reduce

# built sentence from words

words=["My","Name","Is","Gayatri"]
sentence=reduce(lambda i,j : i+" "+j,words )
print(sentence)
```

### initializer in reduce() is the optional third argument that provides the starting value for the reduction operation
###   syntax : reduce(function, iterable, initializer)

```python
from functools import reduce

nums = []
result = reduce(lambda x, y: x + y, nums, 0) # ---> here we give third argument is 0
print(result)
```

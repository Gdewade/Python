# if Statement

### used to check single conditions

### Syntax:
##### if condition :
#####     stat1
#####     stat2

```python
age=int(input("Enter Your Age : "))
print(age)

if age >= 18:
    print("Adult, so you can vote")

print("Programs end")
```

# if - else statement

### used to check 2 conditions

### Syntax :
#### if condition:
####        stat1
#### else:
####         stat2

```python
age=int(input("Enter ur age :"))
print(age)

if age >= 18:
    print("You are eligible for voting")
else:
    print("You are not eligible for voting")

print("Thank You !!")
```

# if-elif-else statement

### used to check multiple conditions

```python
age=int(input("Enter ur age :"))
print(age)

if age>13:
    print("Child")
elif age<18:
    print("Teenager")
else:
    print("Adult")
    
```

### Using Logical operator AND 
##### if both conditon is true then only it will work

```python
age=int(input("Enter ur age: "))
print(age)
nationality=input("Enter ur nationality: ")
print(nationality)

if (age>=18) and (nationality=="indian"):
    print("Eligible for voting")

print("Thank u !!")

```

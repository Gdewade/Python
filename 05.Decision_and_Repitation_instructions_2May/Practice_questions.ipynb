```python
#Q1. What will be the output?

num = 0
while num < 5:
    print("Number:", num)
    num = num + 1
print("Loop finished")

#Output : 

# Number: 0
# Number: 1
# Number: 2
# Number: 3
# Number: 4
# Loop finished
```

```python
#Q2. What will be the output? What is the difference between < and <= in while loop?

# < this operator used when we want to check only less than and <= this operator is used when we want to check less than & equal to
counter = 0
while counter <= 5:
    print("Counter:", counter)
    counter = counter + 1
print("Done")

#Output : 

# Counter: 0
# Counter: 1
# Counter: 2
# Counter: 3
# Counter: 4
# Counter: 5
# Done
```

```python
#Q13. What will be the output? What does continue do here?

#continue keyword is used when we want to skip that variable from that location 

for n in range(10):
    if n==5:
        continue
    print(n)
print("Execution completed")
```

```python
#Q14. What will be the output? Why is 5 missing in the output?

# 5 is missing in output because we used continue keyword for that variable location and this keyword skip that variable value

for digit in range(1,11):
    if digit == 5:
        continue
    print(digit,end=" ")
print()
print("Done")
```

```python
#Q15. What will be the output? What is the difference between break and continue?

#break keyword used to terminate from the block after given condition come in block
#continue keyword used to skip the value of variable after it comes during iteration

for x in range(1,6):
    if x==3:
        break
    print("break loop :",x)

print("\n")

for y in range(1,6):
    if y==3:
        continue
    print("continue loop :",y)
```

```python
#Q19. What will be the output? When does a while loop become infinite?

# when condition never becomes false , then while loop becomes infinite
# for ex;
# i = 1
# while i <= 5:
#     print(i)

attempt=1
while attempt<=3:
    print("Attempt :",attempt)
    attempt=attempt +1

print("All attempts done")
```

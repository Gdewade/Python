## Using without 'with' keyword

# 1. Write () mode
#### when we try to add again using 'w' it removes or overlap already exist data so we used append to add more data again 'a'

```python
# write() : used to write content in file (create new file or open existing file)

msg="Good Morning"

fobj=open("text1.txt",'w')
fobj.write(msg)

fobj.close()
```

```python
# writelines() : used when we want to write multiple lines using list

pets=['Cat\n','Dog\n']

fobj1=open("text2.txt",'w')
content1=fobj1.writelines(pets)
fobj1.close()

print(content1)
```

# 2. Read() mode

```python
# read() : used to read content from file

fobj=open("text1.txt",'r')
content=fobj.read()
fobj.close()
print(content)
```

```python
# read(n) : used to read characters 

fobj2=open("text2.txt",'r')
content2=fobj2.read(2)
fobj2.close()

print(content2)
```

```python
# readline() : used to read cmplt line with \n

fobj3=open("file1.txt",'r')
content3=fobj3.readline()
fobj3.close()

print(content3)
```

```python
# readlines() : used to read all lines of file

fobj4=open("file1.txt",'r')
content4=fobj4.readlines()
fobj4.close()

print(content4)
```

# 3. Append : used to add content in file

```python
msg1="\nRecently completed Bachloers in Computer Science"

fobj5=open("file1.txt",'a')
content5=fobj5.write(msg1)
fobj5.close()

print(content5)

```

## using 'with' keyword : to prevent risk like data loss,errors 

```python
with open("file1.txt",'r') as fobj:
    content=fobj.read()
    print(content)

# Auto close on block exit doesnt need to used close
# if exception occurs, file closes
```

```python
with open("file2.txt",'w') as fobj1:
    msg1="Gayatri Bhau Dewade"
    content1=fobj1.write(msg1)
    print(content1)
```

```python
with open("file2.txt",'a') as fobj2:
    msg="\nFrom Pune"
    content2=fobj2.write(msg)
    content3=fobj2.write("\n Good Afternoon")

    print(content2)
    print(content3)
```

## JSON serialization : convert python obj (list,tuple,set,dict) into JSON format

```python
# serialization:

import json

nums=[10,20,30,40,50]

with open("serial1.json",'w+') as fobj:
    json.dump(nums,fobj)
    fobj.seek(0)
    result=json.load(fobj)
    print(result)
```

## JSON deserialization : convert JSON into python obj (list,tuple,set,dict) 

```python
# deserialization:

import json

with open("serial1.json", "r") as fobj:
    result = json.load(fobj)

print(result)
print(type(result))
```

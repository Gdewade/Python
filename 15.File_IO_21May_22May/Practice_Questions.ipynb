```python
# Q26. Write a program that demonstrates the three-step
#      file I/O process:
#      Step 1: open() — file loads into memory buffer (RAM)
#      Step 2: write()/read() — operations on buffer, not disk
#      Step 3: close() — buffer saves to disk, memory released
#      Add a comment at each step explaining what happens.

# STEP 1: open()
# The file is opened and connected to a memory buffer (RAM).
# Python prepares the file for reading/writing operations.

fobj = open("demo.txt", "w")


# STEP 2: write()
# Data is written to the memory buffer first,
# not directly to the disk immediately.

fobj.write("Hello, Good Afternoon")


# STEP 3: close()
# The buffer content is saved permanently to the disk
# and the memory used by the file object is released.

fobj.close()

```

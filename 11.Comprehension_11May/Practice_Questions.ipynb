```python
# Q11. Write a traditional for-loop first, then rewrite the same
#      logic as a list comprehension:
#      Task: Get all multiples of 5 from 1 to 50.

print("Using for loop ")
for i in range(1,51):
    if (i % 5 == 0):
        print(i)

print('\n')

print("Using comprehension")

divisible5=[i for i in range(1,51) if i % 5 == 0]
print(divisible5)
    
```

```python
# Q28. Given the dictionary:
#      scores = {'Alice': 82, 'Bob': 45, 'Carol': 91, 'Dan': 38}
#      Write a dictionary comprehension to keep only students
#      who scored 50 or above.
#      Expected Output: {'Alice': 82, 'Carol': 91}

scores={'Alice':82,'Bob':45,'Carol':91,'Dan':38}
dict2={i:j for i,j in scores.items() if j >= 50}
print(dict2)
```

```python
# Q34. Write the same task using all three comprehension types
#      and observe the difference in output:
#      Task: Process numbers 1 to 5 and compute n**2
#      i)  As a List
#      ii) As a Set
#      iii) As a Dictionary (n → n**2)

print("As a list")
num1=[n**2 for n in range(1,6)]
print(num1)
print('\n')

print("As a set")
num1={n**2 for n in range(1,6)}
print(num1)
print('\n')


print("As a dcitionary")
num1={n:n**2 for n in range(1,6)}
print(num1)

```

### Difference between for loop and comprehension

```python
# for loop
squares=[]
for n in range(1,6):
    squares.append(n**2)
print(squares)
```

```python
# comprehension syntax 

squares=[n**2 for n in range(1,6)]
print(squares)
```

### 1 . List comprehension

```python
cubes=[n**3 for n in range(1,6)]
print(cubes)
```

```python
sqrs=[n**2 for n in range(1,21) if n% 2 ==0]
print(sqrs)
```

```python
animal=["cat","dog","fish","bird"]
upper=[i.upper() for i in animal if len(i)>3]
print(upper)
```

```python
upper1=[]
for i in animal:
    if len(i)>3:
        upper1.append(i)
print(upper1)
```

```python
lengths=[i for i in animal if i[0]=="c"]
print(lengths)

lengths1=[i for i in animal if i.startswith("c")]
print(lengths1)
```

### 2. Set Comprehension = like list but all values are unique,unorderd, used {} 

```python
set1={i**2  for i in range(1,6)}
print(set1)
print(type(set1))
```

```python
# to remove duplicates 

text="misspooooiiiiaaaa"
unq_set={i for i in text}
print(unq_set)

```

### 3 . Dictionary Comprehension

```python
# syntax : {key- expression: val-expression for item in sequence if condition  }

sq_dict={n:n**2 for n in range(1,6)}
print(sq_dict)
```

```python
animal=["cat","dog","fish","bird"]
new={i:len(i) for i in animal }
print(new)
```

```python
stu={"akshay":59,"rahul":78,"rajesh":73}
marks={i:j for i,j in stu.items() if (j>70)}
print(marks)
```

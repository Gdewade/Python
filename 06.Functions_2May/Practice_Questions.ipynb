```python
#Q10. What will be the output if user enters 50 and 30? What does return allow us to do with the result?

print("Progam Started")

def find_max(num1,num2):
    if num1>num2:
        return num1
    else:
        return num2       #return used to display output and we can access it at anywhere in our program
    
a=int(input("Enter first number : "))
b=int(input("Enter second number : "))

biggest=find_max(a,b)
print("Biggest number :",biggest)
```

```python
#Q11. What will be the output if user enters 40 and 10? Can a function return more than one value?

#Yes function can return more than one value
print("Program started")

def cal_add_sub(a,b):
    addition=a+b
    subtraction=a-b
    return addition,subtraction

n1=int(input("Enter first no. "))
n2=int(input("Enter second no."))

res1,res2=cal_add_sub(n1,n2)

print("Additon :",res1)
print("Subtraction :",res2)
```

```python
#Q12. What will be the output if user enters 6 and 4? How many values does this function return?

print("Program started")

def all_operations(x,y):
    add= x+y
    sub= x-y
    multi= x*y
    div= x/y
    return add,sub,multi,div # this function returns 4 values

n1=int(input("Enter first no. "))
n2=int(input("Enter second no."))

res1,res2,res3,res4=all_operations(n1,n2)

print("Additon :",res1)
print("Subtraction :",res2)
print("Multiplication :",res3)
print("Division :",res4)

print("Program end")
```

```python
#Q13. What will be the output? What is the function naming convention used in Python?

#funcntion nameing conventions are function name should be always return in lower case and when we used multiple words as a function name used _ 

def cal_area(length,breadth):
    area= length * breadth
    return area

def cal_perimeter(length,breadth):
    perimeter=2*(length+breadth)
    return perimeter
l=int(input("Enter the length:"))
b=int(input("Enter the breadth:"))

print("Area :",cal_area(l,b))
print("Perimeter :",cal_perimeter(l,b))
```

```python
# Q15. What will be the output? What are the 5 advantages of using functions?

# Advantges of using functions 
# (1) Once we created function, we can used it multiple times in program
# (2) improve readability
# (3) easy to maintain
# (4) testing and debugging
# (5) modularity 

def calc_simple_interest(principal, rate, time):
    si = (principal * rate * time) / 100
    return si

def calc_compound_details(principal, si):
    total = principal + si
    return total

p = int(input("Enter principal: "))
r = int(input("Enter rate: "))
t = int(input("Enter time: "))

interest = calc_simple_interest(p, r, t)
total_amount = calc_compound_details(p, interest)

print("Simple Interest:", interest)
print("Total Amount:", total_amount)
```

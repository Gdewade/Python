```python
def fun_add_sub(a,b):
    add=a+b
    sub=a-b
    return add,sub # used to return result outside the function

n1=int(input("Enter num1 :"))
n2=int(input("Enter num2 :"))

result1,result2=fun_add_sub(n1,n2) # here result is show by return() keyword

print("Result of addition: ",result1)
print("Result of subtraction: ",result2)

```

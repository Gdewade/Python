```python
#Q15. What will be the output? Compare all three methods for the same data.

dept = "HR"
emp = "Ramesh"
salary = 45000

print("Normal  :", "Dept:", dept, "Emp:", emp, "Salary:", salary)
print("format():", "Dept: {} Emp: {} Salary: {}".format(dept, emp, salary))
print(f"f-string : Dept: {dept} Emp: {emp} Salary: {salary}")

#Output : Normal  : Dept: HR Emp: Ramesh Salary: 45000
#         format(): Dept: HR Emp: Ramesh Salary: 45000
#         f-string : Dept: HR Emp: Ramesh Salary: 45000

# Normal format = we used (,) for printing the data
# .format() =  we used {} as a empty string at will be fill later by variables which are declare in .format()
# f-string =  we declared variables in {} for printing data
```

### -> String Concatenation

```python
roll_no=str(100)
address="pune"

print("My address is "+address+" and my roll_no is "+roll_no)
print("My address is",address,"and my roll_no is",roll_no)
```

### -> format() method

```python
roll_no=str(100)
address="pune"
pin_code=str(411015)

print("My address is {} and my roll_no is {} and pin_code is {}".format(address,roll_no,pin_code))
```

### -> f-String

```python
roll_no=str(100)
address="pune"
pin_code=str(411015)

print(f"My address is {address} and my roll_no is {roll_no} and pin_code is {pin_code}")
```
